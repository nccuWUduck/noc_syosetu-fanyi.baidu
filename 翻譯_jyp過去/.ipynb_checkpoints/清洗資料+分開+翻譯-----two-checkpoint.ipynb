```python
切5000字_list=[]
n=0
for i in japan_list:
    if(len(i)>5000):
        tmp_num=len(i)
        tmp_num=int(tmp_num/5000)+1
        for j in range(tmp_num):
            if(5000*(j+1)>len(i)):
                print(5000*j,len(i)-1)
                break
            print(5000*j,5000*(j+1))
        print("----------",n)
        n+=1
        continue
    print("==========",n)
    n+=1
```

```python
切5000字_list=[]
n=0
限制=1000
for i in japan_list:
    if(len(i)>限制):
        tmp_num=len(i)
        tmp_num=int(tmp_num/限制)+1
        for j in range(tmp_num):
            if(限制*(j+1)>len(i)):
                切5000字_list.append(i[限制*j:len(i)-1])
               # print(5000*j,len(i)-1)
                break
            切5000字_list.append(i[限制*j : 限制*(j+1)])
          #  print(5000*j,5000*(j+1))
        
        print("----------",n)
        n+=1
        continue
    切5000字_list.append(i)
    #print("==========",n)
    n+=1
```
