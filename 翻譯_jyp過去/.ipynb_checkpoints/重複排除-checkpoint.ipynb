```python
org_name='異世界生活支援は、信頼cheatでエロlifeを！'
path = 'C:\\Users\\user\\Desktop\\JYPUTER_試做\\' + org_name
```

```python
import os
filename_list=[]
for filename in os.listdir(path):
    print(filename)
    filename_list.append(filename)
```

```python
text=[]
for name in filename_list:
    text_tmp=''
    if(name!='異世界生活支援は、信頼cheatでエロlifeを！__all.txt'):
        with open(os.path.join(path, name), 'r',encoding='UTF-8') as f:
            text_tmp=f.read()
            text.append(text_tmp)
            
```

```python
text_new=[]
for i in text:
    cut_list=[]
    compared1=i[:20]
    print(compared1)
    print('--------------------------\n')
    for word_len in range(len(i)):
        if(i[word_len]==compared1[0]):
#            print("in__if")
            if(i[word_len:word_len+20]==compared1):
#                print("in_sec_if")
                cut_list.append(word_len)
                print(cut_list)
    if(len(cut_list)>1):
        text_new.append(i[:cut_list[1]])
    else:
        text_new.append(i)
```

```python
path=r'C:/Users/user/Desktop/JYPUTER_試做/異世界生活支援は_信頼cheat_11_8_晚上重複排除/'
count_int=0
for i in text_new:
#    print(type(count))
    if not os.path.isdir(path):
        os.makedirs(path)
    count=count_int
    count="%03d"%count
    name='信頼cheat__'+str(count)
    with open(os.path.join(path, name)+'.txt', 'w',encoding="utf-8") as f:
        f.write(i)
    count_int+=1
```
