```python
n_all_file=0
for file in allFileList:
    if (file.endswith(".csv")):
            print(file,n_all_file)
            n_all_file+=1
```

```python
tmp_japan_list=[]
import csv
n=0
for i in range(n_all_file-1):
    name=yourPath+'\\'+'幼馴染'+str(n)+'.csv'
    with open(name,newline='',encoding="utf-8") as csvfile:
      # 讀取 CSV 檔內容，將每一列轉成一個 dictionary
        rows = csv.reader(csvfile)
        for row  in rows:
            tmp_japan_list.append(''.join(row))
    n+=1
```
